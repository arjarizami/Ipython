```python
import matplotlib.pyplot as plt
import time

fib = 0

a = 0

b = 1

c = 0
totaltime = 0
arreglo = []

while fib < 500000:
    iniciotiempo = time.time()
    c=a+b
    a=b
    b=c
    fintiempo= time.time()
    totaltime += fintiempo-iniciotiempo
    arreglo.append(totaltime)
    fib+=1

plt.plot(arreglo)
plt.show()
print 'El numero Fibonacci deseado es:' + str(fib)
```
